```python
import os
import numpy as np
from astropy.table import Table
```

```python
cmip6 = np.load('/Users/chahah/data/noah/cmip6_data.npy') # Nmodels x Nmonths x Nlats x Nlons
lats = np.load('/Users/chahah/data/noah/cmip6_lat.npy') # latitudes
lons = np.load('/Users/chahah/data/noah/cmip6_lon.npy') # longitudes
```

```python
cmip6 *=  30 * 86400 # convert to mm
```

```python
forecast = Table()
forecast['latitude']  = np.empty(cmip6.shape[2] * cmip6.shape[3])
forecast['longitude'] = np.empty(cmip6.shape[2] * cmip6.shape[3])
```

```python
for year in [2022, 2030, 2035, 2040, 2052]: 
    for month in np.arange(1, 13): 
        forecast['precip.%i.%i' % (year, month)] = np.empty((cmip6.shape[2] * cmip6.shape[3], cmip6.shape[0]))
```

```python
for i_lat, lat in enumerate(lats): 
    for i_lon, lon in enumerate(lons): 
        forecast['latitude'][i_lat * cmip6.shape[3] + i_lon] = lat
        forecast['longitude'][i_lat * cmip6.shape[3] + i_lon] = lon
        
        for year in [2022, 2030, 2035, 2040, 2052]: 
            for month in np.arange(1, 13): 
                forecast['precip.%i.%i' % (year, month)][i_lat * cmip6.shape[3] + i_lon] = cmip6[:, (year - 1850) * 12 + month -1, i_lat, i_lon]
```
