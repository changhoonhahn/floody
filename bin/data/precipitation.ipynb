```python
import os
import numpy as np
from astropy.table import Table
```

```python
# precipitation today
rainfall = Table.read('/Users/chahah/data/noah/prism_zipcode_20240911.csv', format='csv')
```

```python
rainfall.rename_column('ZCTA5CE20', 'ZIPCODE')
for col in ['GEOID20', 'CLASSFP20', 'MTFCC20', 'FUNCSTAT20', 'ALAND20', 'AWATER20', 'INTPTLAT20', 'INTPTLON20', 'ORIG_FID']: 
    rainfall.remove_column(col)
```

rename columns

```python
all_columns = list(rainfall.columns)
for col in all_columns: 
    if col[0] == 'm': 
        i_month = int(col[1:])
        
        year = 1990 + (i_month-1) // 12
        month = ((i_month-1) % 12) + 1
        rainfall.rename_column(col, 'precip.%i.%i' % (year, month))
```

```python
for year in np.arange(1990, 2023): 
    precip_year = np.zeros((len(rainfall), 12))
    
    for i, month in enumerate(np.arange(1, 13)): 
        precip_year[:,i] = rainfall['precip.%i.%i' % (year, month)]
        rainfall.remove_column('precip.%i.%i' % (year, month))
        
    rainfall['precip.%i' % year] = precip_year
```
