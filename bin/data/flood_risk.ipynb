# preprocess flood risk data from the First Street Foundation
The dataset from FSF is poorly labeled: PUT URL TO FSF dataset HERE

```python
import os
import numpy as np
from astropy.table import Table
```

```python
fsf = Table.read('/Users/chahah/data/noah/fsf/Flood Risk Data.csv')
```

```python
floodrisk = Table()
floodrisk['ZIPCODE'] = fsf['ZIP_CODE']
```

## average risk score currently (2022)

```python
floodrisk['avg_risk.2022'] = fsf['avg_risk_s']
```

## percent of properties flooded in different scenarios for 2022 and 2052

```python
floodrisk['risk5.2022'] = fsf['pct_fs_ris'] # Percent of properties flooded in the First Street return period 5 scenario (currently)
floodrisk['risk5.2052'] = fsf['pct_fs_r_1'] # Percent of properties flooded in the First Street return period 5 scenario (in 30 years)

floodrisk['risk100.2022'] = fsf['pct_fs_r_2'] # Percent of properties flooded in the First Street return period 100 scenario (currently)
floodrisk['risk100.2052'] = fsf['pct_fs_r_3'] # Percent of properties flooded in the First Street return period 100 scenario (in 30 years)

floodrisk['risk500.2022'] = fsf['pct_fs_r_4'] # Percent of properties flooded in the First Street return period 500 scenario (currently)
floodrisk['risk500.2052'] = fsf['pct_fs_r_5'] # Percent of properties flooded in the First Street return period 500 scenario (currently)
```
